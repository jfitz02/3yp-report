```python
#here we will be getting football.csv rugby.csv and tennis.csv
#using the words in the csv files to generate a bag of words
#then we will be using the bag of words to train a classifier
#to classify tweets as football, rugby or tennis


import pandas as pd
import numpy as np

#read in the csv files, naming single column 'text'
football = pd.read_csv('football.csv', names=['text'])
rugby = pd.read_csv('rugby.csv', names=['text'])
tennis = pd.read_csv('tennis.csv', names=['text'])
esports = pd.read_csv('esports.csv', names=['text'])

football = list(football['text'])
rugby = list(rugby['text'])
tennis = list(tennis['text'])
esports = list(esports['text'])

football[:5]
```

```python
import sqlite3 as sql

#connect to the database
conn = sql.connect('../twitter.db')

#read a random sample of 25 tweets from the database, that are given topic "sport"
tweets = pd.read_sql_query("SELECT tweet_text FROM tweets WHERE tweet_topic='sports' ORDER BY RANDOM() LIMIT 25", conn)

#close the connection
conn.close()

tweets.head()
```

```python
#for each of football, rugby and tennis, we will count the number of times each word appears in the tweets
import re

for tweet in tweets['tweet_text']:
    #convert to lowercase
    tweet = tweet.lower()
    #remove punctuation
    tweet = re.sub(r'[^\w\s]','',tweet)
    print(tweet)
    football_count = 0
    rugby_count = 0
    tennis_count = 0
    esports_count = 0
    for word in football:
        if word in tweet:
            football_count += 1
    for word in rugby:
        if word in tweet:
            rugby_count += 1
    for word in tennis:
        if word in tweet:
            tennis_count += 1
    for word in esports:
        if word in tweet:
            esports_count += 1

    #print topic with highest count
    if rugby_count >= football_count and rugby_count >= tennis_count and rugby_count >= esports_count:
        print('rugby')
    elif football_count >= rugby_count and football_count >= tennis_count and football_count >= esports_count:
        print('football')
    elif tennis_count >= football_count and tennis_count >= rugby_count and tennis_count >= esports_count:
        print('tennis')
    elif esports_count >= football_count and esports_count >= rugby_count and esports_count >= tennis_count:
        print('esports')
```

```python
#given to arrays, calculate the cosine similarity between them
def cosine_similarity(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)

#write a test function to test the cosine similarity function
def test_cosine_similarity():
    a = np.array([1, 2, 3])
    b = np.array([4, 5, 6])
    res = cosine_similarity(a, b)
    print(res)
    assert res == 32 / np.sqrt(14 * 77)
```
